```python
import requests
from bs4 import BeautifulSoup as bs

link = 'https://inshorts.com/en/read'
req = requests.get(link)

soup = bs(req.content, 'html5lib')
box = soup.findAll('div', attrs = {'class':'news-card z-depth-1'})

for i in range(len(box)):
    print('HEADING : ', box[i].find('span', attrs = {'itemprop':'headline'}).text)
    print('TIME : ', box[i].find('span', attrs = {'class':'time'})['content'][:10])
    
    img = box[i].find('div', attrs = {'class':'news-card-image'})
    img_link = img['style'].split("'")[1]
    display(Image(url = img_link, width = 300))
    
    print('BODY : ', box[i].find('div', attrs = {'itemprop':'articleBody'}).text, end='\n\n')
    try:
        print('Read More : ', box[i].find('a', attrs = {'class':'source'})['href'])
    except:
        pass
    print('_'*100, end='\n\n')
```
