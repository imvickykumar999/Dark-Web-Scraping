```python
import requests
from bs4 import BeautifulSoup as bs

x = 1
line = []
box = 1
while box:
  link = f'https://github.com/imvickykumar999?page={x}&tab=followers'
  x += 1
  req = requests.get(link)

  soup = bs(req.content, 'html5lib')
  box = soup.findAll('p', attrs = {'class':'color-fg-muted text-small mb-0'})

  for i in box:
    line.append([i for i in i.text.replace('\n', '').split(' ') if i][-1].replace(',', '').replace('.', '').replace('@', ''))

to_insert = list(set(line))
to_insert
```

```python
# importing the required libraries
import gspread
import pandas as pd
from oauth2client.service_account import ServiceAccountCredentials
```

```python
# define the scope
scope = ['https://spreadsheets.google.com/feeds',
         'https://www.googleapis.com/auth/drive',
        ]

jfile = 'client_secrets.json'

# add credentials to the account
creds = ServiceAccountCredentials.from_json_keyfile_name(jfile, scope)

# authorize the clientsheet 
client = gspread.authorize(creds)
client
```

```python
url = 'https://docs.google.com/spreadsheets/d/1ZXGn0kdgCfGg6m5x9B2vRnhE4lLr6SxrIHCSH5qiKw4/edit?usp=sharing'

sheet = client.open_by_url(url) 
sheet
```
