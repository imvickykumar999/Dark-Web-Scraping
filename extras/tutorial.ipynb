```python
def inform(username):
    response = requests.get(f'https://www.instagram.com/{username}/?__a=1').json()
    graph = response['graphql']['user']
    
    value = [
        graph['username'],
        graph['full_name'], 
        graph['edge_followed_by']['count'], 
        graph['edge_follow']['count'], 
        graph['is_private'], 
        graph['profile_pic_url_hd'], ]

    key = ['username', 'full_name', 'followers', 'following', 
           'is_private', 'profile_pic_url_hd', ]

    insta = dict(zip(key, value))
    print(graph['biography'])
    return insta
```
